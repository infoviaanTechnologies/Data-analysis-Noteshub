```python
import pandas as pd
import numpy as np
```

```python

data = {
    'Name': ['Alice', 'Bob', np.nan, 'Dave', 'Eve', 'Bob'],
    'Age': [25, np.nan, 35, 40, np.nan, 30],
    'Gender': ['Female', 'Male', 'Male', 'male', 'Female', 'Male'],
    'Salary': [50000, 60000, 55000, 65000, 70000, 60000]
}

df = pd.DataFrame(data)
df
```

```python
#1. cheack null/ blank data/None/np.nan
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.notnull()
```

```python
df.notnull().sum()
```

```python
#Remove Rows with Missing Data dropna -> drop nan
df.dropna(inplace=True)
df
```

#### Repalce Missing Values

```python

data = {
    'Name': ['Alice', 'Bob', np.nan, 'Dave', 'Eve', 'Bob'],
    'Age': [25, np.nan, 35, 40, np.nan, 30],
    'Gender': ['Female', 'Male', 'Male', 'male', 'Female', 'Male'],
    'Salary': [50000, 60000, 55000, 65000, 70000, 60000]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
#df.fillna()  --> use fill missing values
#df.fillna(0)   --> its fill 0 for all dataFrame
```

```python
df['Name'].fillna("unknown",inplace=True)
```

```python
df
```

```python
age_mean=df['Age'].mean()
print(age_mean)
```

```python
df['Age'].fillna(age_mean,inplace=True)
df
```

#### Remove Duplicate rows

```python

data = {
    'Name': ['   Alice', 'Bob', 'Bob', 'Dave', 'Eve', 'Bob'],
    'Age': [25, 23, 23, 40, 34, 23],
    'Gender': ['Female', 'Male', 'Male', 'male', 'Female', 'Male'],
    'Salary': [50000, 60000, 60000, 65000, 70000, 60000]
}

df = pd.DataFrame(data)
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
##replace
#df['Gender'] = df['Gender'].replace({'male': 'Male'})
```

```python
df
```

```python
df['Gender'] = df['Gender'].replace({'Male':'M','Female':'F','male':'M'})
df
```

```python
#rename column /index 
df.rename(columns={'Salary': 'MonthlySalary'}, inplace=True) 
df
```

```python
df.rename(index={0: 'Emp1', 1: 'Emp2'}, inplace=True)
df
```

```python
#
df['Age'] = df['Age'].astype(float)
df
```

```python
#Work with Text Columns (String Operations)
#Convert to lowercase
df['Name'] = df['Name'].str.lower()
df['Name'] = df['Name'].str.upper()


df
```

```python
#Strip unwanted spaces
df['Name'] = df['Name'].str.strip()
df
```

```python
#Find string length
df['Name_Length'] = df['Name'].str.len()
```
