```python
import numpy.random as rd
```

```python
#random value 0 to 1
ar=rd.rand(100)
print(ar)
```

```python
ar=rd.rand(10)
print(ar)
```

```python
ar_2d=rd.rand(2,3)
print(ar_2d)
```

```python
ar_3d =rd.rand(4,2,3)
print(ar_3d)
```

```python
# randn()

ar=rd.randn(100)
print(ar)
```

```python
ar=rd.randn(100)
print(ar)
```

```python
#ranf()

ar=rd.ranf(5)
print(ar)
```

```python
#randint()

ar=rd.randint(1,100,size=5)
print(ar)
```

```python
#2d array
ar=rd.randint(1,1000,size=(2,3))
print(ar)
```

```python
#3rd array using randint()
ar=rd.randint(1,100,size=(4,2,3))
print(ar)
```

```python
#as a argument dimension
ar= rd.rand(1,2,3,4)
print(ar)
```

```python
ar=rd.ranf(size=(4,2))
print(ar)
```
